```python
import numpy as np
from numpy.random import random as srand
import matplotlib.pyplot as plt
from scipy.stats import mode
import json

def plot_results(scores, probs, title=""):
    fig, axs = plt.subplots(1, 2, figsize=(15, 5))
    fig.suptitle(title)
    for i in range(scores.shape[1]):
        axs[0].plot(range(scores.shape[0]), scores[:, i])
    
    for i in range(probs.shape[1]):
        axs[1].scatter(range(probs.shape[0]), probs[:,i,0])
    plt.show()
```

```python
scores = np.load("data/data-score.npy")
probs = np.load("data/data-probs.npy")
```

```python
plot_results(scores, probs)
```
